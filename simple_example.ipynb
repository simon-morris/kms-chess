```python
import kms_chess
import itertools
import pandas as pd
```

## Download files from Lichess

```python
kms_chess.download_game_files(1)
```

* Read the first 50 games in the file
* get the material distribution at each point past the filter in "get_details_from game"
* view as a pandas DataFrame

```python
rows = []
for game in itertools.islice(kms_chess.games_gen(), 50):
    rows += kms_chess.get_details_from_game(game, 45)
pd.DataFrame(rows, columns = ["Event", "Result", "WhiteElo", "BlackElo", "game_id", "moves"] + kms_chess.PIECE_SYMBOLS)
```
